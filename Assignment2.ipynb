# Q20 Heat Transfer

```python
# Heat Transfer(Q) = mass(m) * Specific Heat Capacity(c) * Change in temperature(delta_t)
# Given values for heat transfer = 500J, mass= 8kg, Initial tempereture=20C, final temperature=30

m = 8
c = 500
Initial_t = 20
Final_t = 30
delta_t= Final_t-Initial_t
Q = m * c * delta_t
print("Heat Transfered:",Q,"Joules")
```
